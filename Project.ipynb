```python
import pandas as pd
import numpy as np
from matplotlib.pyplot import show
import seaborn as sns
```

```python
df=pd.read_csv("C:/Users/Sakshi Gupta/Downloads/Cars93.csv")
df
```

```python
# show models having price more than 25
df[['Model','Price']][df['Price']>25].sort_values(by='Price')
```

```python
df.shape
```

```python
df.info()
```

```python
#show price of type large
df[['Type','Price']] [df['Type']=='Large'].sort_values(by='Type')
```

```python
#show length of car for USA origin 
df[['Origin','Length']] [df['Origin']=='USA']
```

```python
#show model where price between 20 to 30
df[['Model','Price']] [(df['Price']>20) & (df['Price']<30)]
```

```python
# show avg price of all types
df.groupby('Type')['Price'].mean()
```

```python
#show hiegest enginesize of each model
df.groupby('Model')[['EngineSize']].max().sort_values(by='EngineSize',ascending=False)
```

```python
#shoe lowest horsepowe of each type of car
df.groupby('Type')[['Horsepower']].min()
```
